**FOR**

```
for i in range(5):
  print(i) 
  print("merhaba")
```

```
a = [1, 2, 25, 0 , 100]

for el in a:
  print(el)
```

```
for i in range(25, 50):
  print(i)
```

**WHILE**

```
sayi = 10


while sayi < 20:
  print(sayi)
  
  sayi = sayi + 1 
```

```
20 < 20 
```

**BREAK**

```
while True:
  print("merhaba")
  break 
  
  
for i in range(1000):
  break 
  print("merhaba")
```

**CONTINUE**

```
for i in range(5):
  if i == 3:
    continue
    print(i)
  else:
    print(i)
```

**PASS**

```
sayi = 60

if sayi > 50:
  pass 

for i in range(5):
  pass
```

```
for i in range(50):
  for j in range(50):
    print(i, j)
```
